```python
known_image = face_recognition.load_image_file("dummy.jpg")
unknown_image = face_recognition.load_image_file("test.jpg")

shubham_encoding = face_recognition.face_encodings(known_image)[0]
unknown_encoding = face_recognition.face_encodings(unknown_image)[0]

results = face_recognition.compare_faces([shubham_encoding], unknown_encoding)
```

```python
print(results)
```

```python
score = face_recognition.face_distance([shubham_encoding], unknown_encoding)
print(score)
```
